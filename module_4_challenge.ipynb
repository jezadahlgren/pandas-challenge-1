## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['unit_total'] = df['unit_price'] * df['qty']
df.head(3)
check_unit_total = df[["unit_price", "qty", "unit_total"]]
check_unit_total.head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

df['total_weight'] = df['qty'] * df['unit_weight']

def shipping_price (total_weight):
    if total_weight > 50:
        return total_weight * 7
    return total_weight * 10
df['shipping_price'] = df['total_weight'].apply(shipping_price)

check_col_shipping = df[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]]
check_col_shipping.head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
sales_tax_rate = 1.0925
df['total_price'] = (df['unit_total'] + df['shipping_price']) * sales_tax_rate

df['total_price'] = df['total_price'].round(2)
df[['unit_total' , 'shipping_price' , 'total_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head(3)

# Notes: below is my individual check
# check_line_cost = ["unit_cost", "shipping_price", "line_cost"]
# df[check_line_cost] = df[check_line_cost].round(2)
# check_line_cost_df = df[check_line_cost]
# check_line_cost_df.head(3) 
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['line_profit'] = df['total_price'] - df['line_cost']
df.head(3)

# below is the individual columns I am checking
#df[['line_cost' , 'total_price' , 'line_profit']].head()
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Note for this there are 5 client_IDs with the highest qty: [24741, 38378, 66037, 46820, 33615]

client_df = pd.DataFrame()  # Create an empty DataFrame to store the results

# Background equations, the sums will appear after my for client_id.unique statement
# client_id, unit_cost, qty -- are all columns in the original data

client_df['client_id'] = df['client_id']
client_df['unit_cost'] = df['unit_cost']
client_df['qty'] = df['qty']


sales_tax_rate = 1.0925  # hold for the sales tax

# adding basic equations 
df['unit_total'] = df['unit_price'] * df['qty'] # Calculate the individual price by qty
df['total_weight'] = df['qty'] * df['unit_weight']  # Calculate total weight for shipping price

def shipping_price_new(total_weight):  # creation of the shipping price function
    if total_weight > 50:
        return total_weight * 7
    return total_weight * 10

df['shipping_price'] = df['total_weight'].apply(shipping_price_new)  # additional column (shipping_price) not a sum

# Calculating the main metrics 
df['total_price'] = (df['unit_total'] + df['shipping_price']) * sales_tax_rate # calc for total_price
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price'] # cal for line_cost
df['line_profit'] = df['total_price'] - df['line_cost']

client_series = df['client_id'].value_counts()  # to create a series where each unique client_id is my index
# clients_df = pd.DataFrame(client_series.sort_values(ascending=False))  # basic sort can delete

client_data = []  # to create a list to stick my new variables into

# I attempted to add the full calculation for each but it did not work so I 
# calculate the sum for each unique client and add it to my function below
for client_id in df['client_id'].unique():
    total_qty = df.loc[df['client_id'] == client_id, 'qty'].sum()  # sum of qty
    total_shipping_price = df.loc[df['client_id'] == client_id, 'shipping_price'].sum()  # sum of shipping
    total_line_cost = df.loc[df['client_id'] == client_id, 'line_cost'].sum()  # must be after shipping_price_new function ('Total Cost')
    total_line_price = df.loc[df['client_id'] == client_id, 'total_price'].sum()  # this is my total_price equation
    total_line_profit = df.loc[df['client_id'] == client_id, 'line_profit'].sum()  # total profit code

    client_data.append({
        'Client ID': client_id,
        'Client Count': client_series[client_id], 
        'Units': total_qty, 
        'Shipping': total_shipping_price, 
        'Total Revenue': total_line_price, 
        'Total Cost': total_line_cost,
        'Total Profit': total_line_profit
        })

client_df = pd.DataFrame(client_data)
client_five = client_df.sort_values(by=['Client Count'], ascending=[False])
client_five_top = client_five.head(5).reset_index(drop=True) 
client_five_top_df = client_five_top.copy()
client_five_top_df = client_five_top.sort_values(by=['Total Profit'], ascending=[False])
client_five_top_df

#[24741, 38378, 66037, 46820, 33615] - top five clients ids
```
